```python
import scipy.spatial.distance
from geopy.distance import geodesic
mat = scipy.spatial.distance.cdist(location_df[["latitude", "longitude"]], location_df[["latitude", "longitude"]], lambda u, v: geodesic(u,v).mi)
miles_distances = pd.DataFrame(mat, index=location_df['state'])
miles_distances.columns = location_df['state']
#miles_distances.replace(0, np.nan,inplace=True)
miles_distances
```

```python
test_df = location_df.copy()
new_miles_distances = miles_distances.copy()
for i in range(5):
    for row in new_miles_distances.values:
        row[np.nanargmin(row)] = np.NaN
        new_close = np.where(new_miles_distances.values == np.nanmin(new_miles_distances.values, axis=1)[:,None],new_miles_distances.columns,False)
        #print([[x for x in i if bool(x)][0] for i in new_close])
        test_df['Closest State #' + str(i + 1)] = [[x for x in i if bool(x)][0] for i in new_close]
    dist_to_state = []
    for j in np.arange(len(test_df)):
        dist_to_state = dist_to_state + [miles_distances.loc[test_df['state'][j],test_df['Closest State #' + str(i + 1)][j]]]
    test_df['Distance to Closest State #' + str(i + 1)] = dist_to_state

#miles_distances.loc[test_df.loc[0,['state']][0],test_df.loc[0,['Closest State #1']][0]]
test_df
#new_miles_distances.values
```
