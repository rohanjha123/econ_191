```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly as pl
import seaborn as sns
%matplotlib inline
```

```python
bls_df = pd.read_excel("monthly-listing-3.xlsx",header=1)
bls_df["NAICS_col"] = bls_df["Industry code[1]"].astype(str).str[:2]
bls_df["start_year"] = pd.DatetimeIndex(bls_df["Work stoppage beginning date"]).year
bls_df["end_year"] = pd.DatetimeIndex(bls_df["Work stoppage ending date"]).year
bls_df = bls_df[bls_df['end_year'] >= 1993]
bls_df.head()
```

```python
bls_df_subset = bls_df[~bls_df['States'].isna()]
#bls_df_subset = bls_df_subset[~bls_df_subset['States'].str.contains(",")]
bls_df_subset
```

```python
testing_df = bls_df_subset.copy()
for index, row in testing_df.iterrows():
    if "," in row['States']:
        states_list = row['States'].split(",")
        testing_df.loc[index, 'States'] = states_list[0]
        if index == 3:
            print(states_list)
        for i in np.arange(1,len(states_list)):
            row_2 = row
            row_2['States'] = states_list[i]
            testing_df.loc[testing_df.index.max()+1, :] = row_2
bls_df_subset = testing_df
bls_df_subset
```

```python
bls_df_subset['num_years'] = bls_df_subset["end_year"] - bls_df_subset["start_year"] + 1
bls_df_subset['num_years'].sum()
```

```python
testing_df = bls_df_subset.copy()
for index, row in testing_df.iterrows():
    if row['num_years'] > 1:
        i = row['num_years']-1
        start_year = row['start_year']
        while i != 0:
            row_2 = row
            row_2['start_year'] = start_year + i
            testing_df.loc[testing_df.index.max()+1, :] = row_2
            i = i - 1
bls_df_subset = testing_df
bls_df_subset
```

```python
grouped_df = bls_df_subset.groupby(['States','start_year'])['Number of workers[2]'].sum().reset_index()
grouped_df#.head()
```
